```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def perform_linear_regression(x, y, input_value=None):
    # Scatter plot of the data
    plt.scatter(x, y)
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.title("Data Distribution")
    plt.show()

    # Calculate the coefficients (slope and intercept) for the linear regression
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    num = 0
    deno = 0

    for i in range(len(x)):
        num += (x[i] - x_mean) * (y[i] - y_mean)
        deno += (x[i] - x_mean) ** 2

    coeff = num / deno
    print("COEFFICIENT =", coeff)

    intercept = y_mean - (coeff * x_mean)
    print("INTERCEPT =", intercept)

    # Calculate the predicted values
    y_pred = coeff * x + intercept

    # Scatter plot of the data points
    plt.scatter(x, y, c="red", label="Data")

    # Plot the linear regression line
    plt.plot(x, y_pred, label="Linear Regression")

    # If an input value is provided, make a prediction
    if input_value is not None:
        prediction = coeff * input_value + intercept
        print(f"Prediction for input value {input_value} is {prediction}")
        plt.scatter(input_value, prediction, c="green", label="Prediction")

    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.legend()
    plt.title("Linear Regression")
    plt.show()

# Load the data from a CSV file
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Extract the longitude and latitude as features
x = df['longitude'].values
y = df['latitude'].values

# Perform linear regression and make a prediction for a specific input value (e.g., input_value = -122.0)
input_value = -122.0
perform_linear_regression(x, y, input_value)
```
