
A.Instead of directly giving values, randomly generate x and y using random module and do linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random


random.seed(42)
num_samples = 22
x = np.array([random.uniform(1, 50) for _ in range(num_samples)]).reshape((-1, 1))
y = np.array([2 * val + random.uniform(-5, 5) for val in x])

plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()


SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)

mse = mean_squared_error(y, pred)
plt.plot(x, pred, color='green', marker='*')

print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```
